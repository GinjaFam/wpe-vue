```python
file_path = '/workspace/data_files/CN_table.csv'

raw_cn = pd.read_csv(file_path, delimiter=';', encoding='utf-8', header=0)
# bring to lower case
raw_cn.columns = raw_cn.columns.str.lower()
# remove spaces
raw_cn.columns = raw_cn.columns.str.strip()
# print(raw_cn.columns)
# print(raw_cn.head())
# raw_cn.set_index('') 
display(raw_cn.head())
raw_cn.info()

#return the unique 'cover treatment' values for the c
```

```python
# grab a column values and return a series object
raw_cn['cover type']
#grab unique values
# raw_cn['cover type'].unique()
# grab the third element from the cover type column - zero indexed
# raw_cn['cover type'][3]
# can also grab multiple columns, must be a list!
cols = ['a','b']
raw_cn[cols]
```

```python
# for series (one column) i can simply access the row data by index

raw_cn['cover type'][2]

# for datafarames i cannot do that, but i need to use something like .loc

raw_cn[cols].loc[2]
# Can also be written like this
raw_cn.loc[2, cols]

# i can also just use indexes (for both columns and rows) by using iloc. Can also slice it

raw_cn.iloc[:,[0,1,-1,-2,-3,-4]]
```

## Renaming

```python
display(raw_cn)
raw_cn.columns

# change columns names

raw_cn.columns = ['l_type', 'treatment', 'h_condition', 'hsg_a', 'hsg_b', 'hsg_c',
       'hsg_d']

# could also just rename one column

# raw_cn.rename({"culo":"type"}, axis=1, inplace=True)
# raw_cn.columns
raw_cn.info
```

```python
raw_cn.info()
display(raw_cn)
```

```python
# To check the number of values missing per rows
raw_cn.isna().sum(axis=1)

# Then i can return just the rows that have null values inside, by using an operator

raw_cn[raw_cn.isna().sum(axis=1)>0]

# there are two options: fix or throw away

# To drop rows with nan , can change the axis to get rid of the columns
# raw_cn.dropna()

# i will fill the missing data with the string'---'

df = raw_cn.fillna('---')
```
